```python
from IPython.core.display import HTML
import pandas as pd
problems = pd.read_csv('vpexp1-problems.csv.gz', compression='gzip')
solutions_chunks = pd.read_csv('vpexp1-solutions.csv.gz', compression='gzip', iterator=True, chunksize=1000)
solutions = pd.concat(list(solutions_chunks), ignore_index=True)
data = pd.merge(solutions, problems, left_on='problem-argshash', right_on='argshash')
print data
```

```python
by_algo = data.groupby(['pack', 'select', 'itemsort', 'binsort'])['failcount'].agg(
    { 'num_items_unpacked' : sum, 'num_problems_solved' : lambda x: sum(1 for y in x if y == 0) }).sort(columns=['num_items_unpacked'])
HTML(by_algo.iloc[:20].to_html() + "<p>...</p>" + by_algo.iloc[-20:].to_html())
```

```python
output = ''
for col_name in ['num_dimensions', 'num_items', 'slack', 'cov', 'split']:
    output += '<h2>' + col_name + '</h2>'
    for value, group in data.groupby(col_name):
        output += '<h3>' + col_name + ' = ' + str(value) + '</h3>'
        by_algo = data[data[col_name] == value].groupby(['pack', 'select', 'itemsort', 'binsort'])
        pct_items_unpacked = by_algo.apply(lambda x: 100.0 * sum(x['failcount']) / sum(x['num_items']))
        num_problems_solved = by_algo.apply(lambda x: sum(1 for y in x['failcount'] if y == 0))
        temp_df = pd.DataFrame({ 'pct_items_unpacked' : pct_items_unpacked, 'num_problems_solved' : num_problems_solved}).sort(
                      columns=['pct_items_unpacked', 'num_problems_solved'], ascending=[True, False])
        output += temp_df.iloc[:20].to_html()
        output += '<p>...</p>'
        output += temp_df.iloc[-20:].to_html()
HTML(output)
```

```python
results = solutions.group(["pack", "select", "itemsort", "binsort"], {}, { 'problems_solved' : 0, 'items_unmapped' : 0 }, 
    'function(curr, result) { if (curr.failcount == 0) { result.problems_solved++; }; result.items_unmapped += curr.failcount; }')
```

```python
# problem-wise aggregation...
results = solutions.group(['problem-argshash', 'num_dimensions', 'num_bins', 'num_items', 'slack', 'cov', 'split'], {}, { 'solver_count' : 0, 'solvers' : [] },
                          'function(curr, result) { if (curr.failcount == 0) { result.solver_count++; result.solvers.push(curr.pack ' +
                          '+ "_" + curr.select + "_" + curr.itemsort + "_" + curr.binsort); } }')

for result in sorted(results, key=lambda x: x['solver_count']):
    print "%(problem-argshash)s\t%(num_dimensions)d\t%(num_bins)d\t%(num_items)d\t%(slack)f\t%(cov)f\t%(split)d\t%(solver_count)d" % result
    if result['solver_count'] < 100:
        print result['solvers']
```

```python
results = solutions.group(['num_dimensions'], {}, 
                          { 'chosen_fails' : 0, 'avg_fails' : 0, 'bf_fails' : 0, 'bfl2_fails' : 0 },
                          '''
                          function(curr, result) { 
                              if (curr.pack == 'pack_by_bins' && curr.select == 'a:maxratio' && 
                                  curr.itemsort == 'd:sum' && curr.binsort == 'd:maxratio') { 
                                  result.chosen_fails += curr.failcount;
                              }
                              if (curr.pack == 'pack_by_bins' && curr.select == 'a:sum' && 
                                  curr.itemsort == 'd:sum' && curr.binsort == 'a:sum') { 
                                  result.bf_fails += curr.failcount;
                              }
                              if (curr.pack == 'pack_by_bins' && curr.select == 'a:lnorm' && 
                                  curr.itemsort == 'd:lnorm' && curr.binsort == 'a:lnorm') { 
                                  result.bfl2_fails += curr.failcount;
                              }
                              result.avg_fails += 1.0 * curr.failcount / 3630.0;
                          }''')

for result in sorted(results, key=lambda x: x['num_dimensions']):
    print "%(num_dimensions)d\t%(chosen_fails)d\t%(bf_fails)d\t%(bfl2_fails)d\t%(avg_fails)f" % result 
```

```python
results.sort(key=lambda x: x['num_dimensions'])
results_df = DataFrame(results)
print results_df
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(results_df['num_dimensions'], results_df['chosen_fails'], label='chosen fails')
ax.plot(results_df['num_dimensions'], results_df['bf_fails'], label='bf fails')
ax.plot(results_df['num_dimensions'], results_df['bfl2_fails'], label='bf l2 fails')
plt.xlim(min(results_df['num_dimensions']), max(results_df['num_dimensions']))
plt.legend()
plt.show()
```
