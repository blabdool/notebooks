```python
from IPython.core.display import HTML
import pandas as pd
problems = pd.read_csv('vpexp1-problems.csv.gz', compression='gzip')
solutions_chunks = pd.read_csv('vpexp1-solutions.csv.gz', compression='gzip', iterator=True, chunksize=1000)
solutions = pd.concat(list(solutions_chunks), ignore_index=True)
data = pd.merge(solutions, problems, left_on='problem-argshash', right_on='argshash')
print data
```

```python
algorithm_variables = ['pack', 'select', 'itemsort', 'binsort']
data_by_algo = data.groupby(algorithm_variables)
algorithm_ranking = data_by_algo['failcount'].agg(
    { 'num_items_unpacked' : sum, 'num_problems_solved' : lambda x: sum(1 for y in x if y == 0) }).sort(
    columns=['num_items_unpacked', 'num_problems_solved'], ascending=[True, False])
HTML(algorithm_ranking.iloc[:250].to_html() + "<p>...</p>" + algorithm_ranking.iloc[-250:].to_html())
```

```python
favored_algo_index = ('pack_by_bins', 'a:maxratio', 'd:sum', 'd:maxratio')

independent_variables = ['num_dimensions', 'num_items', 'slack', 'cov', 'split']

output = ''
for variable in independent_variables:
    output += '<h2>Independent Variable: ' + variable + '</h2>'
    for value, group in data.groupby(variable):
        output += '<h3>' + variable + ' = ' + str(value) + '</h3>'
        valgroup = group.groupby(algorithm_variables)
        temp_df = pd.DataFrame({ 'pct_items_unpacked' : valgroup.apply(lambda x: 100.0 * sum(x['failcount']) / sum(x['num_items'])),
                                 'pct_problems_solved' : valgroup.apply(lambda x: 100.0 * sum(1 for y in x['failcount'] if y == 0) / len(x))})
        favored_pct_items_unpacked = temp_df.loc[favored_algo_index]['pct_items_unpacked']
        output_df = temp_df[temp_df['pct_items_unpacked'] < favored_pct_items_unpacked].sort(columns='pct_items_unpacked')
        output += '<p>' + str(favored_algo_index) + ' percent items unpacked: ' + str(favored_pct_items_unpacked)
        output += ' algorithms doing better: ' + str(len(output_df)) + '</p>'
        if len(output_df) > 0:
            output += output_df.to_html()
HTML(output)
```

```python
import matplotlib.pyplot as plt

algo_indexes = [('pack_by_bins', 'a:maxratio', 'd:sum', 'd:maxratio'),
                ('pack_by_bins', 'a:maxdiff', 'none', 'd:maxratio')]
#                ('pack_by_bins', 'a:maxratio', 'd:lnorm', 'd:maxratio')]

for variable in independent_variables:
    temp_df = pd.DataFrame(dict((index, group.groupby(variable).apply(lambda x: 100.0 * sum(x['failcount']) / sum(x['num_items']))) 
                                for index, group in data_by_algo))
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_title('percent items unpacked vs. ' + variable)
    ax.set_xlabel(variable)
    ax.set_ylabel('percent items unpacked')
    for algo_index in algo_indexes:
        algo_df = temp_df[algo_index]
        ax.plot(algo_df.index, algo_df, label=str(algo_index))
    avg_df = temp_df.apply(lambda x: sum(x) / len(x), axis=1)
    ax.plot(avg_df.index, avg_df, label='average')
    #min_df = temp_df.apply(min, axis=1)
    #ax.plot(min_df.index, min_df, label='min')
    plt.xlim(min(temp_df.index), max(temp_df.index))
    plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1)) 
    plt.show()
```
