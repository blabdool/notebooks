```python
from pymongo import MongoClient
client = MongoClient()
db = client.vpexp1
solutions = db.solutions
solutions.count()
```

```python
results = solutions.group(["pack", "select"], {}, { 'problems_solved' : 0, 'items_unmapped' : 0 }, 
    'function(curr, result) { if (curr.failcount == 0) { result.problems_solved++; }; result.items_unmapped += curr.failcount; }')
```

```python
print solutions.count() / (2 * 15)
print
print "solved\tpack\tselect"
for result in sorted(results, key=lambda x: x['problems_solved'], reverse=True):
    print "%(problems_solved)d\t%(pack)s\t%(select)s" % result
print
print "unmapped\tpack\tselect"
for result in sorted(results, key=lambda x: x['items_unmapped']):
    print "%(items_unmapped)d\t%(pack)s\t%(select)s" % result
```

```python
results = solutions.group(["pack", "select", "num_dimensions"], {}, { 'problems_solved' : 0, 'items_unmapped' : 0 }, 
    'function(curr, result) { if (curr.failcount == 0) { result.problems_solved++; }; result.items_unmapped += curr.failcount; }')
```

```python
print "dims\tsolved\tpack\tselect"
for result in sorted(results, key=lambda x: (x['num_dimensions'], x['problems_solved']), reverse=True):
    print "%(num_dimensions)d\t%(problems_solved)d\t%(pack)s\t%(select)s" % result
print
print "dims\tunmapped\tpack\tselect"
for result in sorted(results, key=lambda x: (x['num_dimensions'], x['items_unmapped'])):
    print "%(num_dimensions)d\t%(items_unmapped)d\t%(pack)s\t%(select)s" % result
```

```python
results = solutions.group(["pack", "select", "itemsort", "binsort"], {}, { 'problems_solved' : 0, 'items_unmapped' : 0 }, 
    'function(curr, result) { if (curr.failcount == 0) { result.problems_solved++; }; result.items_unmapped += curr.failcount; }')
```

```python
print solutions.count() / (2 * 15 * 11 * 11)
print
print "solved\tpack\tselect\titemsort\tbinsort"
for result in sorted(results, key=lambda x: x['problems_solved'], reverse=True)[:20]:
    print "%(problems_solved)d\t%(pack)s\t%(select)s\t%(itemsort)s\t%(binsort)s" % result
print
print "unmapped\tpack\tselect\titemsort\tbinsort"
for result in sorted(results, key=lambda x: x['items_unmapped'])[:20]:
    print "%(items_unmapped)d\t%(pack)s\t%(select)s\t%(itemsort)s\t%(binsort)s" % result
```

```python
for num_dimensions in [1, 2, 3, 4]:
    print "num_dimensions: %d\n" % (num_dimensions)
    results = solutions.group(['pack', 'select', 'itemsort', 'binsort'], { 'num_dimensions' : num_dimensions }, 
                              { 'problems_solved' : 0, 'items_unmapped' : 0 }, 
        'function(curr, result) { if (curr.failcount == 0) { result.problems_solved++; }; result.items_unmapped += curr.failcount; }')
    
    print "solved\tpack\tselect\titemsort\tbinsort"
    for result in sorted(results, key=lambda x: x['problems_solved'], reverse=True)[:40]:
        print "%(problems_solved)d\t%(pack)s\t%(select)s\t%(itemsort)s\t%(binsort)s" % result
    print
    print "unmapped\tpack\tselect\titemsort\tbinsort"
    for result in sorted(results, key=lambda x: x['items_unmapped'])[:40]:
        print "%(items_unmapped)d\t%(pack)s\t%(select)s\t%(itemsort)s\t%(binsort)s" % result
    print
```

```python
# problem-wise aggregation...
results = solutions.group(['problem-argshash', 'num_dimensions', 'num_bins', 'num_items', 'slack', 'cov', 'split'], {}, { 'solver_count' : 0, 'solvers' : [] },
                          'function(curr, result) { if (curr.failcount == 0) { result.solver_count++; result.solvers.push(curr.pack ' +
                          '+ "_" + curr.select + "_" + curr.itemsort + "_" + curr.binsort); } }')

for result in sorted(results, key=lambda x: x['solver_count']):
    print "%(problem-argshash)s\t%(num_dimensions)d\t%(num_bins)d\t%(num_items)d\t%(slack)f\t%(cov)f\t%(split)d\t%(solver_count)d" % result
    if result['solver_count'] < 100:
        print result['solvers']
```
